# Helical Motion in a Magnetic Field

In class we derived the equations of motion for a charged particle in a constant magnetic field:
$$
x(t) = x_0 + \frac{v_{0y}}{\omega} \left( 1 - \cos \omega t \right)
$$
$$
y(t) = \frac{v_{0y}}{\omega} \sin\omega t
$$
$$
z(t) = v_{0z} t
$$
Let's plot its motion!

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

omega = 2*np.pi  # so the period is 1 s
v0y = 10
v0z = 1
x0 = 0
t = np.linspace(0, 10, 1000)
x = x0 + v0y / omega * (1 - np.cos(omega*t))
y = v0y / omega * np.sin(omega*t)
z = v0z * t

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
