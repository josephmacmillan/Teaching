We're going to focus on plotting in this tutorial.  Here's some code that plots a simple function:

```python
import numpy as np
import matplotlib.pyplot as plt

def simple_function(x):
    return x*np.sin(x)

# make some arrays to hold data
x = np.linspace(0, 30, 100)
f = simple_function(x)

# a quick plot
plt.plot(x, f)
```

$$
v_t = g \tau.
