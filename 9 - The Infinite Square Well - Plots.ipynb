# The Infinite Square Well

Plotting the wave functions and their squares.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def V(x):
    a = 1
    return np.piecewise(x, (x<0, x > 0, x>a), (100, 0, 100))

def Phi(x, n = 1):
    a = 1
    return np.piecewise(x, (x<0, x > 0, x > a), (0, lambda x: np.sin(n * np.pi * x), 0))

def E(n = 1):
    return n**2

N = 1000

x = np.linspace(-1, 2, N)

fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(1, 2, 1)
ax.axis('off')
ax.plot(x, V(x), color='black', lw=2)
ax.set_xlim(-0.2, 1.2)
ax.set_ylim(-1, 12)

# plot the wave functions
for n in range(1, 4):
    p = Phi(x, n)
    ax.plot(x, p + E(n))
    ax.axhline(E(n), color="black", lw=1)

# and the squares
ax = fig.add_subplot(1, 2, 2)
ax.axis('off')
ax.plot(x, V(x), color='black', lw=2)
ax.set_xlim(-0.2, 1.2)
ax.set_ylim(-1, 12)
for n in range(1, 4):
    p = Phi(x, n)
    p2 = np.abs(p)**2
    ax.plot(x, p2 + E(n))
    ax.axhline(E(n), color="black", lw=1)
    
    
plt.savefig("fig_infinite_square_well.pdf")
    
```

```python
# and plot the high-n wave function
p = np.abs(Phi(x, 20))**2

plt.plot(x, V(x), color='black')
plt.plot(x, p)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
plt.xlabel(r'$x / a$')
plt.ylabel(r'$|\Phi_{20}|^2$')
plt.savefig("fig_isw_large_n.pdf")
```
