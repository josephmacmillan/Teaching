# Solving Laplace's Equation
## Via Separation of Variables

So in class we saw that the solution to Laplace's equation to the situation shown below was
$$
V(x,y) = \frac{4 V_0}{\pi} \sum_{n\text{ odd}}^{\infty} \frac{1}{n} e^{-n\pi x / a} \sin \left( \frac{n\pi y}{a} \right).
$$
What exactly does this mean?  Well, let's try plotting up the solution.




```python
import matplotlib.pyplot as plt
import numpy as np
```

First, I'll set $x = 0$ in the expression above so we can look at the $y$ dependence.

```python
V0 = 1.0
a = 1.0
N = 100

y = np.linspace(0, a, N)
Vy = np.zeros(N)

fig = plt.figure(figsize=(10, 6))

for n in range(1, 100, 2):
    Vy += 4.0 * V0 / n / np.pi * np.sin(n * np.pi * y / a)
    
ax = fig.add_subplot(1,1,1)
ax.plot(y, Vy)
```

So, by adding more and more terms to the sum, we can get closer and closer to the constant potential $V_0$ that exists at $x = 0$.

What about the $x$ dependence as well?  For that we'll need a 3D plot.

```python
x = np.linspace(0, a, N)
x, y = np.meshgrid(x, y)

V = 2.0 * V0 / np.pi * np.arctan(np.sin(np.pi * y / a) / np.sinh(np.pi * x / a))

#print(V)
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, V)
ax.set_xlabel('x')
ax.set_ylabel('y')
#ax.set_zlabel('V')
```
